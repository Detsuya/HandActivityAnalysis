```python
%matplotlib widget

import cv2
import vapoursynth as vs
from vapoursynth import core
import numpy as np
import time
import scipy
from scipy import ndimage
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable


def show_video(frames, original=None):
    if len(frames) > 1:
        output = np.zeros((1000, 1), np.uint8)
        if len(frames) % 2:
            frames.append(0)
        new_weight = round(1920 / len(frames) * 2)
        new_height = round(1000 / 2)
        for i in range(0, len(frames), 2):
            output = np.concatenate((output, np.concatenate(
                (cv2.resize(frames[i], (new_weight, new_height), interpolation=cv2.INTER_AREA),
                 cv2.resize(frames[i + 1], (new_weight, new_height), interpolation=cv2.INTER_AREA)), axis=0)),
                                    axis=1)
    else:
        output = frames[0]
    cv2.imshow('Difference', output)
    if original:
        cv2.imshow('Original', original)


def get_mean(video, frame_num, frame_ptr):
    frames = []
    for i in range(frame_num):
        frame = video.get_frame(frame_ptr)
        frame_ptr += 1
        frame = np.asarray(frame[0])
        frames.append(frame)
    # calculate the mean
    mean_frame = np.mean(frames, axis=0).astype(np.single)
    mean_frame = scipy.ndimage.gaussian_filter(mean_frame, sigma=5, truncate=4)
    return mean_frame


def get_masks(video, baseTime, bendingTime):
    base_frame = get_mean(video, 5, baseTime)
    frame = video.get_frame(bendingTime)
    curr_frame = np.asarray(frame[0])
    curr_frame = scipy.ndimage.gaussian_filter(curr_frame, sigma=5, truncate=4)
    curr_frame = np.subtract(curr_frame, base_frame)
    curr_frame = np.divide(curr_frame, base_frame)
    minMask = (curr_frame < np.percentile(curr_frame, 5)).astype(np.single)
    maxMask = ((curr_frame > np.percentile(curr_frame, 95)).astype(np.single))*2
    minMask = scipy.ndimage.binary_opening(minMask, np.ones((25, 25)))
    minMask = scipy.ndimage.binary_closing(minMask, np.ones((25, 25)))
    maxMask = scipy.ndimage.binary_opening(maxMask, np.ones((25, 25)))
    maxMask = scipy.ndimage.binary_closing(maxMask, np.ones((25, 25)))
    return (minMask, maxMask)
    
    
video = core.ffms2.Source(source='D:/VideoTest/45_2.mp4', format=vs.RGBS, width=3088, height=2064)
plt.rcParams.update({'figure.max_open_warning': 0})
```
